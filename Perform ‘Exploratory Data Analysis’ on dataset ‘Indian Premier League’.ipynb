```python
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
# Total number of rows and columns

# Rows containing duplicate data
duplicate_rows_df = finaldata[finaldata.duplicated()]
print("number of duplicate rows:", duplicate_rows_df.shape)
```

```python
#counting rows before droping
finaldata.count()
```

```python
df = finaldata.drop_duplicates()
df
```

```python
df.count()
```

```python
print(df.isnull().sum())  ## finding null values
```

```python
# Dropping the missing values.
df = df.dropna() 
df.count()
```

```python
# After dropping the values
print(df.isnull().sum()) 
```

```python
sns.boxplot(x=df["season"])  ##detecting outliers
```

```python
sns.boxplot(x=df["over"])
```

```python
Q1 = df.quantile(0.25)    ##Try to find total outliers
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df.winner.value_counts().nlargest(40).plot(kind="bar", figsize=(10,5))  ######For team
plt.title("season wise winner")
plt.ylabel("season")
plt.xlabel("winner")
```

```python
# Finding the relations between the variables.
plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(20,6))
ax.scatter(df["winner"], df["season"])
ax.set_xlabel("winner")
ax.set_ylabel("season")
plt.show()
```

```python
df.player_of_match.value_counts().nlargest(40).plot(kind="bar", figsize=(10,5))  ######For player_of_match
plt.title("season wise player")
plt.ylabel("season")
plt.xlabel("player_of_match")
```
